## **Please remove special characters from the name column?**

```python
# Empty list for collecting the special characters
Cleaned_name=[]
#looping through name column and ckecking for special characters
for i in df_Oly["name"]:
  for j in i:
    if(j.isalpha()==True):
      continue
    else:
      if j in Cleaned_name:  # to remove duplicates
        continue
      else:
        Cleaned_name.append(j) # appending the special characters into cleaned_name list
```

## **Please check if the age dataset is skewed or symmetric. Based on the results perform the transformation?**

```python
#Create a new figure and make the size (10,8)
plt.figure(figsize=(10,8))
sns.histplot(x='age',data=df_Oly,kde=True)
# Place the age values at a 90-degree angle.
plt.xticks(rotation= 90)
plt.xlabel('age')
plt.ylabel('count')
plt.title('Figure 1:Distribution of age data')
plt.show()
```

```python
df_Oly['age'].skew()
```

Here age data is moderately right skewed

We are having highly skewed data. If there are too much skewness in the data, then many statistical model don't work properly. So in skewed data, the tail region may act as an outlier for the statistical model and we know that outliers adversely affect the model's performance.

So, for reducing the skewness Transformation is used. 
Transformation can be done by 
Square root(sqrt) transformation, 
Cube root(cbrt) transformation, 
Log transformation, 
Boxcox transormation and many others.

We are applying Log Transformation here.

```python
df_Oly_age_log = df_Oly.copy(deep=True)
df_Oly_age_log['age'] = np.log(df_Oly['age'])
df_Oly_age_log['age'].skew()

```

```python
plt.figure(figsize=(10,8))
sns.histplot(x='age',data=df_Oly_age_log,kde=True)
```

--> After applying the Log Transformation we can see that the skewness of the dataset is no more and the data is normally distributed now.
